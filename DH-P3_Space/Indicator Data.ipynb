```python
EDIbase = r'C:\Users\djhon\OneDrive\Desktop\GW_Homework\Project_1.5-ETL\HD_Datasources\EDI.csv'
HDIbase = r'C:\Users\djhon\OneDrive\Desktop\GW_Homework\Project_1.5-ETL\HD_Datasources\HDI.csv'
UEMbase = r'C:\Users\djhon\OneDrive\Desktop\GW_Homework\Project_1.5-ETL\HD_Datasources\UEM.csv'

EDId = pd.read_csv(EDIbase, encoding='latin-1',header=1)
HDId = pd.read_csv(HDIbase, encoding='latin-1',header=1)
UEMd = pd.read_csv(UEMbase, encoding='latin-1',header=1)
```

```python
EDId = EDId.iloc[:,~EDId.columns.str.contains('^Unnamed')]
EDId['Country'] = EDId['Country'].str.upper()
EDId['Source'] = 'EDI'
EDId1 = EDId[['Source','HDI Rank', 'Country','2015','2016','2017']]
EDI = EDId1.iloc[:-1]
EDI
```

```python
HDId = HDId.iloc[:,~HDId.columns.str.contains('^Unnamed')]
HDId['Country'] = HDId['Country'].str.upper()
HDId['Source'] = 'HDI'
HDId1 = HDId[['Source','HDI Rank', 'Country','2015','2016','2017']]
HDI = HDId1.iloc[:-1]
HDI
```

```python
UEMd = UEMd.iloc[:,~UEMd.columns.str.contains('^Unnamed')]
UEMd['Country'] = UEMd['Country'].str.upper()
UEMd['Source'] = 'UEM'
UEMd1 = UEMd[['Source','HDI Rank', 'Country','2015','2016','2017']]
UEM = UEMd1.iloc[:-2]
UEM
```

```python
iframes = [EDI,HDI,UEM]
ciframes = pd.concat(iframes)
ciframes
```
