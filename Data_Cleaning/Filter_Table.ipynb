```python
import pandas as pd
```

```python
import csv

with open('../Resources/ML_Accuracy_Table.csv') as csvfile:
    # create a list to store results
    users = []

    # read the file
    reader = csv.reader(csvfile)

    # for every line in the csv
    for row in reader:

        #create a user dict
        user = {}

        # fill in the user.
        user["country"] = row[0]
        user["data count"] = row[1]
        user["training"] = row[2]
        user["testing"] = row[3]

        # add this user to the list of users
        users.append(user)
```
