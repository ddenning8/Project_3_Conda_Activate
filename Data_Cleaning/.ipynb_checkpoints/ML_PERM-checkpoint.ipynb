```python
ml_data = pd.read_csv('../Resources/ML_data.csv')
```

```python
ml_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
num_encode = LabelEncoder()
```

```python
ml_data['CLASS_OF_ADMISSION_CODE'] = num_encode.fit_transform(ml_data['CLASS_OF_ADMISSION'])
ml_data['FOREIGN_WORKER_INFO_EDUCATION_CODE'] = num_encode.fit_transform(ml_data['FOREIGN_WORKER_INFO_EDUCATION'])
ml_data['COUNTRY_OF_CITIZENSHIP_CODE'] = num_encode.fit_transform(ml_data['COUNTRY_OF_CITIZENSHIP'])
ml_data['EMPLOYER_STATE_CODE'] = num_encode.fit_transform(ml_data['EMPLOYER_STATE'])
ml_data['DAY_BINS_CODE'] = num_encode.fit_transform(ml_data['DAY_BINS'])
ml_data['WAGE_BINS_CODE'] = num_encode.fit_transform(ml_data['WAGE_BINS'])
```

```python
ml_data.head()
```

```python
ml_data2 = ml_data[[
    'CASE_RECEIVED_DATE_MONTH',
    'WAGE_BINS_CODE',
    'CLASS_OF_ADMISSION_CODE',
    'FOREIGN_WORKER_INFO_EDUCATION_CODE',
    'EMPLOYER_STATE_CODE',
    'COUNTRY_OF_CITIZENSHIP_CODE',
    'DAY_BINS_CODE',
    'CASE_STATUS'
]]
```

```python
ml_data2.head()
```

```python
ml_data2.count()
```

```python
X = ml_data2.drop("CASE_STATUS", axis=1)
y = ml_data2["CASE_STATUS"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=42, stratify=y, test_size=0.20)
```

```python
X_scaler = MinMaxScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
# Step 2: Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Create model and add layers
model = Sequential()
model.add(Dense(units=100, activation='relu', input_dim=7))
model.add(Dense(units=100, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
```

```python
# Compile and fit the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=10,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test_scaled, y_test_categorical, verbose=4)
print(
    f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
encoded_predictions = model.predict_classes(X_test_scaled[:10])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)
```

```python
print(f"Predicted classes: {prediction_labels}")
print(f"Actual Labels: {list(y_test[:10])}")
```
