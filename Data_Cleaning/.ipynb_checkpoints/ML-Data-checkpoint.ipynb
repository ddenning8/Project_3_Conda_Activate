```python
data2 = data[data.CASE_STATUS != 'CERTIFIED-EXPIRED']
data3 = data2[data2.CASE_STATUS != 'WITHDRAWN']
```

```python
data3['CASE_STATUS'].value_counts()
```

```python
data4 = data3[data.WAGE_OFFER_UNIT_OF_PAY_9089 == 'Year']
```

```python
data4['WAGE_OFFER_UNIT_OF_PAY_9089'].value_counts()
```

```python
data4.count()
```

```python
regions = []

for row in data4['EMPLOYER_STATE']:
    if row == "PENNSYLVANIA" and "NEW JERSEY" and "NEW YORK" and "CONNECTICUT" and "RHODE ISLAND" and "MASSACHUSETTS" and "VERMONT" and "NEW HAMPSHIRE" and "MAINE":
        regions.append("NORTHEAST")
    elif row == "NORTH DAKOTA" and "SOUTH DAKOTA" and "NEBRASKA" and "KANSAS" and "MINNESOTA" and "IOWA" and "MISSOURI" and "WISONSIN" and "ILLINOIS" and "INDIANA" and "MICHIGAN" and "OHIO":
        regions.append("MIDWEST")
    elif row == "TEXAS" and "OKLAHOMA" and "ARKANSAS" and "LOUISIANA" and "MISSISSIPPI" and "ALABAMA" and "TENNESSEE" and "GEORGIA" and "FLORIDA" and "KENTUCKY" and "SOUTH CAROLINA" and "NORTH CAROLINA" and "VIRGINIA" and "WEST VIRGINIA" and "MARYLAND" and "DISTRICT OF COLUMBIA" and "DELAWARE":
        regions.append("SOUTH")
    else:
        regions.append("WEST")
        
data4['EMPLOYER_STATE_REGION'] = regions
```
